```python
#1 и 2 задания в одном - общие продажи и юниты по брендам, здесь по сути все ясно, сортируем все по value
total_by_brand = df.groupby('brand').agg({'value': 'sum', 'units': 'sum'}).sort_values(by='value', ascending=False)
print("Total Sales and Units per Brand:")
print(total_by_brand)
total_by_brand.plot(kind='bar', figsize=(14, 6), title='Total Sales and Units per Brand')
plt.tight_layout()
plt.show()
```

```python
# 3- топ 5 брендов по месяцам, заполняем нулем те строчки где пусто в датасете, а то чето ошибка у меня была
top5 = df.groupby('brand').agg({'value': 'sum'}).sort_values(by='value', ascending=False).head(5).index.tolist()
monthly_top5 = df[df['brand'].isin(top5)].groupby(['month', 'brand'])['value'].sum().unstack().fillna(0)
print("Monthly Sales (Top 5 Brands):")
print(monthly_top5)
monthly_top5.plot(marker='o', figsize=(14, 6), title='Monthly Sales (Top 5 Brands)')
plt.tight_layout()
plt.show()
```

```python
#4 - топ-7 брендов по кварталам, примерно такой же принцип, сори дальше лень расписывать там долго
top7 = df.groupby('brand').agg({'value': 'sum'}).sort_values(by='value', ascending=False).head(7).index.tolist()
quarterly_top7 = df[df['brand'].isin(top7)].groupby(['quarter', 'brand'])['value'].sum().unstack().fillna(0)
print("Quarterly Sales (Top 7 Brands):")
print(quarterly_top7)
quarterly_top7.plot(marker='s', figsize=(14, 6), title='Quarterly Sales (Top 7 Brands)')
plt.tight_layout()
plt.show()
```

```python
# 5 - лучшие модели по юнитам за год для топ-10
top10 = df.groupby('brand').agg({'value': 'sum'}).sort_values(by='value', ascending=False).head(10).index.tolist()
best_model_year = df[df['brand'].isin(top10)].groupby(['brand', 'model'])['units'].sum().reset_index()
best_model_year = best_model_year.sort_values(['brand', 'units'], ascending=[True, False]).drop_duplicates('brand')
print("Best models by units (year) for Top 10 Brands:")
print(best_model_year)
best_model_year.plot(x='brand', y='units', kind='bar', figsize=(14, 6), title='Best Models by Units (Year)')
plt.tight_layout()
plt.show()
```

```python
#6 лучшие модели за июнь и октябрь для топ10
months_filtered = df[(df['month'] == 6) | (df['month'] == 10)]
best_model_mo = months_filtered[months_filtered['brand'].isin(top10)].groupby(['brand', 'model'])['units'].sum().reset_index()
best_model_mo = best_model_mo.sort_values(['brand', 'units'], ascending=[True, False]).drop_duplicates('brand')
print("best models (june and october) of top 10 brands")
print(best_model_mo)
best_model_mo.plot(x='brand', y='units', kind='bar', figsize=(14, 6), title='best models (june and october) of top 10 brands')
plt.tight_layout()
plt.show()
```

```python
# 7  лучшие модели топ3 брендов по месяцам с помощью head
top3 = df.groupby('brand').agg({'value': 'sum'}).sort_values(by='value', ascending=False).head(3).index.tolist()
best_models = df[df['brand'].isin(top3)].groupby(['brand', 'model'])['units'].sum().reset_index()
best_models = best_models.sort_values(['brand', 'units'], ascending=[True, False]).drop_duplicates('brand')
model_monthly = df.merge(best_models[['brand', 'model']], on=['brand', 'model'])
model_monthly_summary = model_monthly.groupby(['brand', 'model', 'month'])['value'].sum().unstack().fillna(0)
print("month sales of best models (top3 brands)")
print(model_monthly_summary)
model_monthly_summary.T.plot(figsize=(14, 6), title='month sales of best models (top3 brands)')
plt.tight_layout()
plt.show()
```

```python
#8  топ15 брендов онлайн и оффлайн с помощью head
online = df[df['channel'] == 'Internet Sales'].groupby('brand')['value'].sum().sort_values(ascending=False).head(15)
offline = df[df['channel'] == 'Traditional Sales'].groupby('brand')['value'].sum().sort_values(ascending=False).head(15)
print("Топ 15 брендов - Интернет продажи:")
print(online)
print("Топ 15 брэндов - Обычные продажи:")
print(offline)
online.plot(kind='bar', title='Топ 15 брендов - Интернет продажи:', figsize=(12, 5))
plt.tight_layout()
plt.show()
offline.plot(kind='bar', title='Топ 15 брендов - Обычные продажи:', figsize=(12, 5))
plt.tight_layout()
plt.show()
```

```python
# 9 топ3 модели по сегментам
segmented = df.groupby(['segment', 'model'])['units'].sum().reset_index()
top3_by_segment = segmented.sort_values(['segment', 'units'], ascending=[True, False]).groupby('segment').head(3)
print("Top 3 models by segment:")
print(top3_by_segment)
top3_by_segment.pivot(index='segment', columns='model', values='units').plot(kind='bar', figsize=(14, 6), title='Top 3 Models by Segment')
plt.tight_layout()
plt.show()
```

```python
#10 топ модели по диапазонам цен
ranges = {
    "200<": df[df['value'] < 200_000],
    '200-400': df[(df['value'] >= 200_000) & (df['value'] < 400_000)],
    "400-600": df[(df['value'] >= 400_000) & (df['value'] < 600_000)],
    '600-800': df[(df['value'] >= 600_000) & (df['value'] < 800_000)],
    "800-1000": df[(df['value'] >= 800_000) & (df['value'] < 1_000_000)],
    '>1000': df[df['value'] >= 1_000_000]}
segment_top = {}
for label, frame in ranges.items():
    x = frame.groupby('model').agg( {'units': 'sum', 'value': 'sum'}).sort_values('units', ascending=False).head(1)
    segment_top[label] = x
    print(f"Segment: {label}")
    print(x)
    x.plot(kind='bar', y='units', title=f'Top Model by Units ({label} KZT)', figsize=(8, 5))
    plt.tight_layout()
    plt.show()
```

```python
#14 топ10 магазинов по продажам
by_shop = df.groupby(['shop', 'city'])['value'].sum().reset_index()
top10_shop_year = by_shop.sort_values('value', ascending=False).head(10)
print("топ10 магазов по продажам (1 год)")
print(top10_shop_year)
geolocator = Nominatim(user_agent="geoapi")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
top10_shop_year['location'] = top10_shop_year['city'].apply(lambda x: geocode(f"{x}, Kazakhstan"))
top10_shop_year['lat'] = top10_shop_year['location'].apply(lambda loc: loc.latitude if loc else None)
top10_shop_year['lon'] = top10_shop_year['location'].apply(lambda loc: loc.longitude if loc else None)
geo_df = gpd.GeoDataFrame(top10_shop_year, geometry=gpd.points_from_xy(top10_shop_year.lon, top10_shop_year.lat))
world = gpd.read_file("https://raw.githubusercontent.com/nvkelso/natural-earth-vector/master/geojson/ne_110m_admin_0_countries.geojson")
kz = world[world['ADMIN'] == "Kazakhstan"]
fig, ax = plt.subplots(figsize=(10, 6))
kz.plot(ax=ax, color='lightgrey')
geo_df.plot(ax=ax, color='purple', markersize=50)
plt.title("Топ 10 магазинов по продажам (1 год)")
plt.show()
```
